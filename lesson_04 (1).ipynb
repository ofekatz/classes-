### Part 1: Comparison Operators 
Comparison operators compare two values and return True or False.

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal to | `5 != 3` | `True` |
| `>` | Greater than | `7 > 5` | `True` |
| `<` | Less than | `3 < 5` | `True` |
| `>=` | Greater or equal | `5 >= 5` | `True` |
| `<=` | Less or equal | `4 <= 3` | `False` |

Code cell:


```python
# Try comparison operators
print(10 > 5)        # True
print(10 < 5)        # False
print(10 == 10)      # True
print(10 != 5)       # True
print(10 >= 10)      # True

# Compare strings
print("apple" == "apple")    # True
print("apple" == "Apple")    # False (case sensitive)
```

Important: Use == to compare, not = (which is for assignment)

---
### Part 2: Logical Operators 
Logical operators combine multiple conditions.

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `and` | Both must be True | `True and True` | `True` |
| `or` | At least one True | `True or False` | `True` |
| `not` | Opposite | `not True` | `False` |

Code cell:


```python
# and operator - both must be True
print(True and True)       # True
print(True and False)      # False
print(False and False)     # False

# or operator - at least one must be True
print(True or False)       # True
print(False or False)      # False

# not operator - reverses the result
print(not True)            # False
print(not False)           # True

# Combining with comparisons
age = 16
print(age > 13 and age < 20)    # True (teenager)
print(age < 13 or age > 65)     # False
```

## Part 3: if Statements
The if statement lets your code make decisions.  
Syntax:

```python
if condition:
    # code runs only if condition is True
    print("Condition is True!")
```

Code cell:


```python
# Simple if statement
age = 16

if age >= 18:
    print("You can vote")

if age < 18:
    print("You cannot vote yet")
```

---
### Part 4: if/else Statements 
Use else to run code when the condition is False.  
Code cell:

```python
age = 16

if age >= 18:
    print("You are an adult")
else:
    print("You are a minor")
```

Another example:


```python
temperature = 25

if temperature > 30:
    print("It's hot outside")
else:
    print("It's not too hot")
```

---
### Part 5: if/elif/else Statements 
Use elif (else if) for multiple conditions.  
Code cell:

```python
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")
```

Another example:


```python
time = 14  # 24-hour format

if time < 12:
    print("Good morning")
elif time < 18:
    print("Good afternoon")
else:
    print("Good evening")
```
